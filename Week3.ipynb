# if문

```python
a = 4
if a % 2 == 0:
    b = "짝수"
else:
    b = "홀수"
print(b)
```

```python
a = 5
if a % 2 == 0:
    b = "짝수"
else:
    b = "홀수"
print(b)
```

```python
# 삼항 연산자 사용
a = 4
b = "짝수" if a % 2 == 0 else "홀수"
print(b)
```

```python
# and 조건 : 조건이 모두 참일 때만 참
a, b = 3, 2
if a > 2 and b < 3:
    print("둘 다 참")
```

```python
# or 조건 : 조건이 모두 거짓인 경우에만 거짓
a, b = 1, 2
if a > 2 and b < 3:
    print("둘 다 참")
elif a > 2 or b < 3:
    print("최소한 둘 중 하나 참")
else:
    print("둘 다 거짓")
```

# 리스트 컴프리헨션 실습 예제

```python
# 1 ~ 10까지의 제곱수를 원소로 가지는 리스트
a = [i ** 2 for i in range(1, 11)]
print(a)
```

```python
# 1 ~ 10까지의 제곱수 중 짝수의 제곱수만을 원소로 가지는 리스트
b = [i ** 2 for i in range(2, 11, 2)]
print(b)
```

```python
a = list()
for i in range(1, 11):
    if i % 2 == 0:
        a.append(i ** 2)
print(a)
```

```python
b = [i ** 2 for i in range(1, 11) if i % 2 == 0]
print(b)
```

```python
# 구구단 결과 값을 가지는 원소
numbers = [i * j for i in range(2, 10) for j in range(1, 10)]
print(numbers)
```

```python
numbers = [f"{i} * {j} = {i * j}" for i in range(2, 10) for j in range(1, 10)]
print(numbers)
numbers
```

```python
a = ["바나나", "딸기", "포도", "딸기", "딸기", "바나나"]
b = dict()
for i in a:
    if i not in b:
        b[i] = 1
    else:
        b[i] += 1
print(b)

b = dict()
for i in a:
    b[i] = b.get(i, 0) + 1
print(b)
```

```python
from collections import defaultdict
b = defaultdict(int)
for i in a:
    b[i] += 1
print(b)
```

## sort 함수와 sorted 함수

```python
x = [4,3,1,2]

# 기본 : 오름차순 정렬
y = sorted(x)
print(x,y)
```

```python
x = [4,3,1,2]
x.sort()
print(x)
```

```python
x = [4,3,1,2]

# 내림차순 정렬 : reverse=True
y = sorted(x, reverse=True)
print(x, y)
```

```python
x = [-4, 3, -1, 2]
y = sorted(x, reverse=True)
print(y)
```

```python
x = [-4, 3, -1, 2]

# 절대값 정렬
y = sorted(x, key=abs, reverse=True)
print(y)
```

```python
a = ["바나나", "딸기", "포도", "딸기", "딸기", "바나나"]
b = Counter(a)

# 가장 많이 존재하는 것을 상위 n번째까지 보여달라는 코드
b.most_common(2)
```
