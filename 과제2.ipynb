### 문제 1
`replace_first(lst, value)` 라는 형태의 함수가 있습니다. 이 함수는 list 타입의 변수인 lst와 임의의 값을 가진 변수인 value 를 인자로 받아서 `lst[0]`을 `value`로 함수 내에서 바꾸는 역할을 합니다. 이 함수 `replace_first(lst, value)`를 작성하세요. **호출자의 리스트가 실제로 바뀌는지** 보여주세요.

```python
# Problem 1
# Write your code below.
def replace_first(lst, value):
  lst[0] = value

my_list = [10, 20, 30]
print(f"함수 호출 전: {my_list}")

replace_first(my_list, 999)

print(f"함수 호출 후: {my_list}")
```

### 문제 2
위와 비슷한 함수인 `replace_first_immutable(tpl, value)`가 있습니다. 이 함수는 tuple 타입의 변수인 tpl과 임의의 값을 가진 value 를 인자로 받아서, 첫 원소가 `value`인 **새로운 튜플**을 반환합니다. 이 함수 `replace_first_immutable(tpl, value)`를 작성하세요. (tuple 자료형인 `tpl`은 변경할 수 없으므로 새로운 튜플을 만들어 반환하는 것입니다.)

```python
# Problem 2
# Write your code below.
def replace_first_immutable(tpl, value):
  temp_list = list(tpl)

  temp_list[0] = value

  return tuple(temp_list)

my_tuple = (10, 20, 30)
print(f"함수 호출 전 원본: {my_tuple}")
print(f"함수 호출 전 id: {id(my_tuple)}")

print("-" * 30)

new_tuple = replace_first_immutable(my_tuple, 999)

print(f"함수 호출 후 원본: {my_tuple}")
print(f"함수가 반환한 값: {new_tuple}")
print(f"함수 호출 후 id: {id(my_tuple)}")
print(f"반환된 튜플의 id: {id(new_tuple)}")
```

### 문제 4
`dict.get`을 사용해 `safe_get()`를 구현하세요. `safe_get()` 함수는 `d, key, default` 변수를 매개변수로 사용합니다. `d` 변수는 dictionary, `key` 변수는 찾고 싶은 key, `default` 변수는 `key`가 `d`에 존재하지 않을 때 반환할 값입니다. 단, `safe_get()`함수를 호출 할 때 `dafault`에 인자를 줄 수도 있고 주지 않을 수도 있습니다. 예를 들어, 다음과 같은 결과가 나와야 합니다.

d = {'a': 1}  
print(safe_get(d, 'a'))  
print(safe_get(d, 'missing'))  
print(safe_get(d, 'missing', 0))

==>

1  
None  
0  

```python
# Problem 4
# Write your code below.
def safe_get(d, key, default=None):
  return d.get(key, default)

d = {'a': 1}

print(safe_get(d, 'a'))

print(safe_get(d, 'missing'))

print(safe_get(d, 'missing', 0))
```

### 문제 5
리스트 컴프리헨션을 사용하여, `text` 라는 변수에 저장된 문자열에서 모음 (a,e,i,o,u) 만 골라 리스트로 만드는 프로그램을 코딩하세요. (힌트: 문자열 안의 모든 알파벳 ch 에 대해 if ch in "aeiouAEIOU" 를 검사하세요.)

예: text="Big Apple"  
==> 결과: ['i', 'A', 'e']

```python
# Problem 5
# Write your code below.
text = "Big Apple"

vowels = [ch for ch in text if ch in "aeiouAEIOU"]

print(vowels)
```

### 문제 6
일반 `dict`와 `get`을 사용해 `['apple','apple','banana','cherry','banana','apple']`의 단어별 빈도수를 세고 그 결과를 딕셔너리 출력하세요.

```python
# Problem 6
# Write your code below.
words = ['apple', 'apple', 'banana', 'cherry', 'banana', 'apple']

counts = {}

for word in words:
  counts[word] = counts.get(word, 0) + 1

print(counts)
```

### 문제 7
`collections.defaultdict(int)`로 **문제 6을 다시 구현**하고 딕셔너리를 출력하세요.

```python
# Problem 7
# Write your code below.
from collections import defaultdict

words = ['apple', 'apple', 'banana', 'cherry', 'banana', 'apple']

counts = defaultdict(int)

for word in words:
  counts[word] += 1

print(counts)
```

### 문제 15

### 문제 17 (난이도 상)
`zip`과 `enumerate`를 함께 사용하여 `['a','b','c']`의 각 원소를 그 **인덱스** 및 `[10,20,30]`의 원소와 짝지어 `[('a',0,10), ...]` 형태를 만드세요.

```python
# Problem 17
# Write your code below.
letters = ['a', 'b', 'c']
numbers = [10, 20, 30]

result = [(value, index, number) for (index, value), number in zip(enumerate(letters), numbers)]

print(result)
```
