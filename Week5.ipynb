```python
b = pd.Series([3, 4, 5, 6, 10000000000000000])
b
```

```python
l = [1, 2, 3, 4, 5]
pd.Series(l)
```

```python
# DataFrame 클래스 사용
pd.DataFrame(l)
```

```python
l2 = [3,4,5,6,7]
df = pd.DataFrame(l)
df["새로운 열"] = l2
df
```

```python
# 데이터프레임에서 열의 차원(원소) 개수가 맞지 않으면 오류 발생
l3 = [1,2,3,4,5,6,7]
df["3열"] = l3
df
```

```python
df
```

```python
# 데이터프레임 자료형 확인
type(df)
```

```python
# 데이터프레임 자료형 확인 방법2
df.dtypes
```

```python
# 데이터프레임 열 이름 확인 방법
df.columns
```

```python
# 데이터프레임 열 이름 변경 방법
df.columns = ["x", "y"]
df
```

```python
df.columns
```

```python
# 데이터프레임 열 이름 하나만 변경 방법
# 새로운 데이터프레임을 변환하는 것이기 때문에 원래 데이터프레임에는 영향을 주지 않음
df2 = df.rename(columns={"x" : "z"})
df2
```

## 판다스 심화 활용

```python
score = [['정약용', 70, 80, 90], ['이순신', 30, 100, 90], ['허균', 100, 30, 50]]
df3 = pd.DataFrame(score)
df3
```

```python
df3.columns = ["이름", "국어", "체육", "수학"]
df3
```

```python
df4 = df3.rename(columns={"이름" : "성명"})
df4
```

```python
df5 = df4.rename(columns={"국어" : "Korean", "수학" : "Math"})
df5
```

```python
score_d = {"성명":["정약용","이순신","허균"],"국어":[70,30,100],"체육":[80,100,30],"수학":[90,90,50]}
score_d
```

```python
df6 = pd.DataFrame(score_d)
df6
```

```python
# 데이터프레임의 값 확인 방법
df6.values
```

```python
# 데이터프레임이 몇 행 몇 열의 구조인지 확인하는 방법
df6.shape
```

```python
# 데이터프레임의 인덱스의 개수 및 범위 확인 방법
df6.index
```

```python
# 데이터프레임 열 이름 및 자료형 확인 방법
df6.columns
```

```python
# 데이터프레임 열 이름 및 자료형 확인 방법2
df6.dtypes
```

```python
# 데이터프레임 열 자료형 변경 방법
df7 = df6.astype({"국어" : float, "수학" : str})
df7
```
