```python
class stocks:
    def __init__(self, name, price, count=0):
        self.name = name
        self.price = price
        self.count = count

lg = stocks("LG전자", 150000)
print(lg.name, lg.price, lg.count)
```

```python
class stocks:
    def __init__(self, name, price, count=0):
        self.name = name
        self.price = price
        self.count = count
    def click(self, num_times = 1):
        self.count += num_times


lg = stocks("LG전자", 150000)
print(lg.name, lg.price, lg.count)
```

```python
lg.click()
```

```python
print(lg.name, lg.price, lg.count)
```

```python
lg.click(3)
print(lg.name, lg.price, lg.count)
```

```python
class stocks:
    def __init__(self, name, price, count=0):
        self.name = name
        self.price = price
        self.count = count
    def click(self, num_times = 1):
        self.count += num_times
        self.temp = 10


lg = stocks("LG전자", 150000)
print(lg.name, lg.price, lg.count)
```

```python
print(lg.name, lg.price, lg.count, lg.temp)
```

```python
class stocks:
    def __init__(self, name, price, count=0):
        self.name = name
        self.price = price
        self.count = count
    def __repr__(self):
        return f"{self.name} {self.price} {self.count}"
    def click(self, num_times = 1):
        self.count += num_times
        self.temp = 10


lg = stocks("LG전자", 150000)
print(lg.name, lg.price, lg.count)
```

# 상속

```python
class substocks(stocks):
    def reset(self):
        pass

sublg = substocks("LG전자", 150000)
print(sublg)
```

```python
sublg.click()
print(sublg)
```

```python
class substocks(stocks):
    def reset(self):
        self.count = 0
    def click(self, num_times = 100):
        self.count += num_times

sublg = substocks("LG전자", 150000)
print(sublg)
```

```python
sublg.click()
```

```python
sublg.reset()
print(sublg)
sublg.click()
sublg.click()
sublg.click()
```
