## 파이썬 연산

### 숫자형 연산

```python
a=1
b=2
c, d, e=3, 4, 5
print(a+b, a-b, a*b, a/b)
print(c**b, b**0.5)
print(e/b, e//b, e%b)
```

```python
# 숫자형 포맷팅 2가지 방법
e, f = 3, 4
g = e + f
print("e + f의 값은 {}".format(g))
print(f"e + f의 값은 {g}")
```

### 문자형 연산

```python
c = "hello"
print(c)
```

```python
# 문자열 인덱싱 방법 : 변수명[]
print(c[3])
```

```python
# 문자열 슬라이싱 방법 : 변수명[시작:끝]
# 시작과 끝을 생략하면 각각 변수의 처음과 끝
print(c[:3])
print(c[2:4])
```

## 문자열 내장 함수

### 문자 개수 세기 : count() 함수

```python
single_string="To be, or not to be, that is the question"
single_string.count("t")
```

### 문자 위치 찾기 : find() 함수

```python
single_string="To be, or not to be, that is the question"
single_string.find("t")
```

### 첫 글자만 대문자로 변경하기 : capitalize() 함수

```python
single_string="to be, or not to be, that is the question"
text = single_string
text.capitalize()
```

### 모든 문자를 대소문자로 변경 : upper(), lower() 함수

```python
single_string="To be, or not to be, that is the question"
text = single_string
print(text.upper()) # text 변수 값을 대문자로 출력
print(text.lower()) # text 변수 값을 소문자로 출력
```

### 특정 문자 변경하기 : replace() 함수

```python
single_string="To be, or not to be, that is the question"
text = single_string
text.replace("question", "QUESTION") # text 변수 값 일부를 대문자로 출력
```

### 단어 단위로 쪼개기 : split() 함수

```python
single_string="To be, or not to be, that is the question"
text = single_string
text.split() # text 변수 값을 공백을 기준으로 분할
```

### 문자열의 시작과 끝에서 주어진 문자 삭제 : strip() 함수

```python
"http://www.Python.org".strip("http://")
```

```python
" Welcome to Hanbit.co.kr ".strip()
```
