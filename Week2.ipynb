# 얕은 복사 깊은 복사

```python
a=1
b=a
a=3
print(b)
```

```python
a=[1,2,3]
b=a
a[0] = 999
print(b)
```

```python
a=[1,2,3]
b=a[:]
a[0] = 999
print(b)
```

```python
a=[[1,2],[3,4]]
b=a[:]
a[0][0] = 999
print(b)
```

```python
prices=[85000, 140000, 200000]
copyprices = prices
prices[0] = 99999
print(copyprices)
```

```python
prices=[85000, 140000, 200000]
copyprices = prices[:] # copyprices = prices.copy()
prices[0] = 99999
print(copyprices)
```

```python
prices=[85000, 140000, 200000]
copyprices = prices.copy()
prices[0] = 99999
print(copyprices)
```

```python
prices=[["과자", 85000], ["음료", 140000], ["주류", 200000]]
copyprices = prices[:]
prices[0][1] = 99999
print(copyprices)
```

```python
import copy
prices=[["과자", 85000], ["음료", 140000], ["주류", 200000]]
copyprices = copy.deepcopy(prices) # deepcopy : 완벽하게 다른 곳을 참조하도록 하는 깊은 복사 함수
prices[0][1] = 99999
print(copyprices)
```

# 딕셔너리

```python
a = {}
print(a)
```

```python
a = dict()
print(a)
```

```python
a = {"과자": 85000, "음료": 140000, "주류": 200000}
print(a)
print(a["과자"])
```

```python
a["아이스크림"] = 50000
print(a)
```

```python
# 키 값만 출력하는 메서드
a.keys()
```

```python
# 실제 value만 출력하는 메서드
a.values()
```

```python
# 딕셔너리 키와 값을 튜플 형태로 출력하는 메서드
a.items()
```

```python
for i in a:
    print(i)
```

```python
for i in a.values():
    print(i)
```

```python
for k, v in a.items():
    print(k, v)
```

```python
a.get("과자")
```

```python
print(a.get("주스"))
```

```python
# 딕셔너리에 등록 되어 있지 않을 경우 일정한 값을 출력하는 방법
# 딕셔너리에 해당 키가 있을 경우 그거에 해당하는 값 출력
print(a.get("주스", 100))
print(a.get("주류", 100))
```

```python
l = ["바나나", "딸기", "수박", "딸기", "바나나"]
d = dict()
for i in l:
    d[i] = d.get(i, 0) + 1
print(d)
```

```python
# 간단하게 사용하는 방법
from collections import Counter
Counter(l)
```
