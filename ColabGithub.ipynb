```python
from google.colab import userdata
import os
```

```python
# 1. Colab Secrets에서 토큰 안전하게 불러오기
token = userdata.get("GITHUB_TOKEN")
```

```python
# 2. 사용자 이름과 저장소 이름 설정
username = "mygithub05253"
repo_name = "BigDataProgrammingClass"
```

```python
# 3. 토큰을 사용하여 저장소 주소 만들기
# 이 방식을 사용하면 코드에 토큰이 직접 노출되지 않습니다.
repo_url_with_token = f"https://{token}@github.com/{username}/{repo_name}.git"
```

```python
# 4. 저장소 clone 또는 원격 주소 설정
# 만약 디렉토리가 이미 존재한다면 clone 대신 원격 주소를 업데이트합니다.
if not os.path.exists(repo_name):
  !git clone {repo_url_with_token}
else:
  # 이미 폴더가 있다면 원격 저장소 주소만 안전한 URL로 업데이트
  %cd {repo_name}
  !git remote set-url origin {repo_url_with_token}
  %cd ..
```
