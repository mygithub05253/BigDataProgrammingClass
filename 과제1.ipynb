### 문제 1
%, `.format()`, f-string을 각각 사용하여 **한 번씩** 다음을 출력하세요: `Name: Ada, Age: 36` (주어진 변수: `name='Ada'`, `age=36`).

```python
# Problem 1
# Write your code below.
name='Ada'
age=36
print('Name: %s, Age: %d' % (name, age))
print('Name: {}, Age: {}'.format(name, age))
print(f'Name: {name}, Age: {age}')
```
