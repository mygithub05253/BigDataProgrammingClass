### 문제 6
`a=1; b=a; a=3`을 실행한 뒤 `b`를 출력하세요. 왜 그런 값이 나오는지 **주석으로 설명**하고, 실행 전/후에 `id()`와 `is`로 동일성도 확인하세요. (`id`는 변수가 참조하는 객체의 주소를 반환하는 함수(예: `id(x)`)이고, `is`는 두 변수의 참조 객체 주소가 같은지를 확인하는 연산자(예: `a is b`)이다.)

```python
# Problem 6
# Write your code below.

# b는 a가 참조하는 객체의 메모리 주소를 가리킨다.
a = 1
b = a
print(id(a))
print(id(b))
print(a is b)

# a는 새로운 객체의 메모리 주소를 가리킨다.
# b는 원래의 객체의 메모리 주소를 가리킨다.
# 즉, a의 값 변경이 b에 영향을 주지 않습니다.
a = 3
print(id(a))
print(id(b))
print(a is b)
```

### 문제 8

### 문제 9
`a=[[1,2],[3,4]]; b=a[:] ; a[0][0]=999` 실행 후 `a`와 `b`를 출력하세요. **이번에는 왜 `b`가 바뀌었는지** `id()`와 `is`를 이용하여 주석으로 설명하세요.

```python
# Problem 9
# Write your code below.
a = [[1, 2], [3, 4]]

# b에 a를 얕은 복사 수행
# a[:]는 새로운 최상위 리스트 생성하지만 그 안에 있는 요소들의 메모리 주소만 참조
b = a[:]
print(id(a))
print(id(b))
print(a is b)

print(a)
print(b)

# 내부의 리스트는 a와 b가 공유하고 있는 상태이기 때문에 내부 리스트의 경우에는 동일한 상태
print(id(a[0]))
print(id(b[0]))
print(a[0] is b[0])

# a의 내부 리스트의 요소 변경
# 메모리 주소만 참조하기 때문에 동일하게 나오지 않음
a[0][0] = 999
print(id(a))
print(id(b))
print(a is b)

# a와 b가 공유하고 있는 내부 리스트 객체 내용 수정하는 것이기 때문에 a[0]와 b[0]의 경우에는 동일한 객체 참조
print(id(a[0]))
print(id(b[0]))
print(a[0] is b[0])

print(a)
print(b)
```

### 문제 17 (난이도 상)
**내부 리스트까지 새로 복사**한 **새로운 2차원 리스트**를 반환하는 `deep_copy_2d(lst2d)`를 작성하세요. (`copy` 모듈 사용 금지)

```python
# Problem 17
# Write your code below.
def deep_copy_2d(lst2d):
    new_lst2d = []
    for i in lst2d:
        new_lst2d.append(i[:])
    return new_lst2d

a = [[1,2], [3,4]]
b = deep_copy_2d(a)

print(a)
print(b)

# 최상위 리스트 비교
print(a is b)

# 내부 리스트 비교
print(a[0] is b[0])

a[0][0] = 999
print(a)
print(b)
```

### 문제 18
`a=[[1],[2],[3]]`를 함수에 전달한 뒤, 함수 내부에서 **변경하면 호출자도 바뀌는지**를 시연하여 **인자 전달 방식**을 설명하세요.

```python
# Problem 18
# Write your code below.
def change_list(data):
    print(id(data))
    data[0][0] = 999

a = [[1], [2], [3]]
print(id(a))
print(a)

change_list(a)
print(a)

# 인자 전달 방식 : 객체 참조에 의한 호출 또는 할당에 의한 전달
# a = [[1], [2], [3]] : a라는 변수에 리스트의 주소 저장
# change_list(a) : change_list 함수에 a의 주소를 알려주는 것
# 함수 내부의 data 변수도 동일한 주소를 보유하게 됨
# data[0][0] = 999 : 해당 주소의 값을 999로 변경
# 함수 실행 종료 후 원래 a의 변수 출력 시 함수에서 변경된 내용을 보유하고 있다는 사실 확인 가능
# 객체 자체를 수정한 경우이기 때문에 원래의 변수에 영향을 주게 된 것

def change_list2(data):
    print(id(data))
    data = [[999]]
    print(data)

a = [[1], [2], [3]]
print(id(a))
print(a)

change_list2(a)
print(a)
# data = [[999]]처럼 새로운 리스트를 재할당하는 경우에는 원래 변수에 영향을 주지 않음
```
