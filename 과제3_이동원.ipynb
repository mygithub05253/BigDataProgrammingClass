```
1. 클래스의 기본 구조 (10점)
다음 조건을 만족하는 코드를 작성하시오.

클래스 이름: Student

속성: 이름(name), 학번(student_id), 점수(score)

생성자(__init__)로 세 속성을 초기화

display() 메서드는 이름과 점수를 출력
```

````python
# 학생 정보를 나타내는 클래스
class Student:
    # 생성자 : 이름, 학번, 점수, 속성 초기화 수행
    def __init__(self, name, student_id, score):
        self.name = name
        self.student_id = student_id
        self.score = score

    # display 메서드 : 학생의 이름과 점수 출력
    def display(self):
        print(f"이름: {self.name}, 점수: {self.score}")

# 클래스 인스턴스 생성
student1 = Student("이동원", 202136006, 90)

# display 메서드를 호출해서 정보 출력
student1.display()
````

```
2. 클래스의 상속과 메서드 오버라이딩 (15점)

다음 요구사항을 만족하는 클래스를 작성하시오.

Person 클래스를 만들고 이름과 나이를 속성으로 가진다.

Person의 하위 클래스 Teacher를 만들어, 과목(subject) 속성을 추가한다.

introduce() 메서드를 Teacher에서 오버라이딩하여

"저는 OOO 선생님입니다. 담당 과목은 XXX입니다."를 출력하시오.

두 클래스의 인스턴스를 각각 생성하고 메서드를 실행하시오.
```

````python
# Person 클래스 : 이름과 나이를 속성으로 가짐
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        print(f"저는 {self.name}입니다. 나이는 {self.age}입니다.")

# Teacher 클래스 : Person 클래스를 상속받아 과목 속성을 추가한 클래스
class Teacher(Person):
    def __init__(self, name, age, subject):
        # 부모 클래스의 생성자를 호출해 이름과 나이 초기화
        super().__init__(name, age)

        # 자식 클래스만의 속성인 과목을 추가하는 초기화 수행
        self.subject = subject

    # 부모의 introduce 메서드를 오버라이딩(재정의) 수행
    def introduce(self):
        print(f"저는 {self.name} 선생님입니다. 담당 과목은 {self.subject}입니다.")

# 각 클래스의 인스턴스 생성
person1 = Person("이동원", 24)
teacher1 = Teacher("김선생", 44, "파이썬")

# 각 인스턴스의 introduce() 메서드 수행
person1.introduce()
teacher1.introduce()
````

```
5. 종합 문제 – 클래스 + 판다스 (25점)

Student 클래스를 재사용하여 다음 요구를 충족하시오.

get_grade() 메서드로 점수를 학점(A~F)으로 변환

(90 이상: A / 80~89: B / 70~79: C / 그 외: D)



아래 코드를 실행하고 결과를 확인하시오.



students = [

    Student("Ann", 1001, 92),

    Student("Bob", 1002, 81),

    Student("Chris", 1003, 68)

]



# DataFrame으로 변환

df = pd.DataFrame([vars(s) for s in students])

df["grade"] = [s.get_grade() for s in students]

print(df)



score 평균을 계산하고, 평균 점수를 출력하시오.
```

````python
import pandas as pd

# 1번 문제의 Student 클래스를 재사용하고 get_grade 메서드 추가
# 학생 정보를 나타내는 클래스
class Student:
    # 생성자 : 이름, 학번, 점수, 속성 초기화 수행
    def __init__(self, name, student_id, score):
        self.name = name
        self.student_id = student_id
        self.score = score

    # display 메서드 : 학생의 이름과 점수 출력
    def display(self):
        print(f"이름: {self.name}, 점수: {self.score}")

    # get_grade 메서드 : 점수를 기준으로 학점(A~D)을 반환하는 메서드
    def get_grade(self):
        if self.score >= 90:
            return "A"
        elif self.score >= 80:
            return "B"
        elif self.score >= 70:
            return "C"
        else:
            return "D"

# 학생 인스턴스 리스트 생성
students = [
    Student("Ann", 1001, 92),
    Student("Bob", 1002, 81),
    Student("Chris", 1003, 68)
]

# DataFrame으로 변환

# 각 학생 객체의 속성을 딕셔너리로 변환 후, 데이터프레임 생성
df = pd.DataFrame([vars(s) for s in students])

# 각 학생의 get_grade() 메서드 호출해서 grade 열 추가
df["grade"] = [s.get_grade() for s in students]

print(df)

# score 평균 계산 및 출력
mean_score = df["score"].mean()
print(mean_score)
````

```
6. 보너스 (10점)

Course 클래스를 만들어 학생 인스턴스를 등록하고, 평균 점수를 계산하는 기능을 구현하시오.

```

````python
# Course 클래스 : 과목 정보를 나타내고, 등록된 학생들의 평균 점수를 계산하는 클래스
class Course:
    def __init__(self, course_name):
        self.course_name = course_name
        # 학생들의 객체를 저장할 리스트 생성
        self.students = []

    # register 메서드 : 학생을 과목에 등록하는 메서드
    def register(self, student):
        self.students.append(student)
        print(f"{student.name} 학생이 {self.course_name} 과목에 등록되었습니다.")

    # score_average 메서드 : 등록된 학생들의 평균 점수를 계산하는 메서드
    def score_average(self):
        if not self.students:
            print("등록된 학생이 없습니다.")
            return 0

        # 모든 학생들의 점수를 리스트로 출력
        scores = [s.score for s in self.students]

        # 평균 계산
        average = sum(scores) / len(scores)
        return average

# 학생 인스턴스 생성
student_list = [
    Student("Ann", 1001, 92),
    Student("Bob", 1002, 81),
    Student("Chris", 1003, 68)
]

# Course 인스턴스 생성
python_course = Course("파이썬 교육")

# 반복문을 통해 학생들을 과목에 등록
for student in student_list:
    python_course.register(student)

# 최종 평균 점수 계산 후 출력
student_average = python_course.score_average()
print(student_average)
````
